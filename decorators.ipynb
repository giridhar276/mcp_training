decorators are a way to add extra behavior to a function without chaning original funcitio code

```python
def say_hi():
    print("python")
say_hi()
```

original function = plain gift
decorator = gift wrapper

the gift is still the same but something extra is added to original unfction

```python
def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("after")
    return wrapper
```

```python
@my_decorator
def say_hi():
    print("hi")
```
