```python
# older style   # regular way

def greet(name):
    print(name)

greet("hello")
```

```python
# modern way # using typehint
# the modern recommended way
def greet(name : str)  -> str:
    print(f"Hello {name}")

greet("Gita")
```

```python
def add( a : int , b : int) -> int :
    return a + b

print(add(10,15))
```
